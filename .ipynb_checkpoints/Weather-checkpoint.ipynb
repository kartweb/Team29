```python
import requests
import pandas as pd
```

```python
api_key = "9b1e1cdc1df64891af9154540252910"
lat, lon = 50.879 , 4.700   # Leuven approx
url = f"https://api.weatherapi.com/v1/history.json?key={api_key}&q={lat},{lon}&dt=2025-10-29&interval=15"
resp = requests.get(url)
data = resp.json()
```

```python
hours = data['forecast']['forecastday'][0]['hour']
df = pd.json_normalize(hours)
print(df[['time', 'temp_c', 'wind_kph', 'humidity', 'condition.text']].head())
```
